```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('coachprerakmehta/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa
```

```python
#extracting job posted month 
#df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.month     #here only number of the months will be extraced, not the names
#df_usa['job_posted_month']

#extracting job posted month names
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa['job_posted_month']
```

```python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot
```

```python
#extracting month number
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.sort_values('month_no', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)

#dropping month_no column because we do not need it anymore. We just needed it to sort the months
df_usa_pivot.drop(columns='month_no')
```

```python
df_usa_pivot.plot(kind='line')
```

```python
#we want to plot top 3 jobs
top_3 = df_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()     #we will pass this list into the plot to display the top 3 jobs
top_3
```

```python
#df_usa_pivot[['Data Analyst', 'Data Scientist', 'Data Engineer']].plot(kind='line')    -  we can do like this also
df_usa_pivot[top_3].plot(kind='line')
plt.title('Monthly job postings for top Data Jobs in US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()        #it is used to show what each line, bar, or point in your plot represents(here the top 3 job roles)
plt.show()
```
