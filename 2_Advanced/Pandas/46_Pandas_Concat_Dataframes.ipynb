```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime("%b")
df.job_posted_month
```

```python
# dict_months = {
#     'Jan': df_jan,
#     'Feb': df_feb,
#     'Mar': df_mar,
#     'Apr': df_apr,
#     'May': df_may,
#     'Jun': df_jun,
#     'Jul': df_jul,
#     'Aug': df_aug,
#     'Sep': df_sep,
#     'Oct': df_oct,
#     'Nov': df_nov,
#     'Dec': df_dec
# }

months = df['job_posted_month'].unique()
#For each month in the list months, #This filters the DataFrame and keeps only the rows where job_posted_month is equal to that specific month.
dict_months = {month: df[df['job_posted_month'] == month] for month in months}     
dict_months['Jan']
```

Data for Q1 needed(Jan, Feb & Mar)

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
q1_jobs = df_q1['job_posted_month'].value_counts()
```

```python
q1_jobs.plot(kind='bar')
```
