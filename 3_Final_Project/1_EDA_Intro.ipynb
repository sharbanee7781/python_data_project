Exploratory Data Analysis for all Data Roles

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
import seaborn as sns

#loading data
dataset = load_dataset('coachprerakmehta/data_jobs')
df = dataset['train'].to_pandas()

#data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and pd.notna(x) else x)
#If it's already a list (or is NaN), leave it as-is.
```

```python
#Only get Data Analyst jobs in US
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()
```

```python
sns.set_theme(style='ticks')
 
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)

sns.despine()             #function in Seaborn to remove the top and right borders from a plot
plt.title('Counts of Job Locations for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)
dict_column = {
    'job_work_from_home' : "Work From Home Offered",
    'job_no_degree_mention' : "Degree Requirement",
    'job_health_insurance' : "Health Insurance Offered"
}

for i, (column, title) in enumerate (dict_column.items()):
    #print(i, column, title)
    ax[i].pie(df_DA_US[column].value_counts(), startangle=90, autopct="%1.1f%%", labels=[False, True])     #big value should be labelled first
    ax[i].set_title(title)

fig.tight_layout()
plt.show()
```

```python
df_plot_2 = df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
 
sns.barplot(data=df_plot_2, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)

sns.despine()             #function in Seaborn to remove the top and right borders from a plot
plt.title('Counts of Companies for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel("")
plt.show()
```
