```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
#loading dataset
url = "https://huggingface.co/datasets/coachprerakmehta/data_jobs/resolve/main/data_jobs.csv"
df = pd.read_csv(url)
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
job_counts = df.job_title_short.value_counts()
job_counts
```

```python
job_counts.plot(kind="bar")
```

```python
df[['job_posted_date', 'salary_year_avg']].dropna(subset='salary_year_avg')
```

```python
df.plot(x='job_posted_date', y='salary_year_avg', kind='line')
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df = df.dropna(subset='salary_year_avg')
```

```python
df
```

```python
df = df.sort_values(by='job_posted_month')
```

```python
df[['job_posted_month', 'salary_year_avg']]
```

```python
df.plot(x='job_posted_month', y='salary_year_avg', kind='line')
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.show()
```
