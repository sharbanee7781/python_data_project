```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset
url = "https://huggingface.co/datasets/coachprerakmehta/data_jobs/resolve/main/data_jobs.csv"
df = pd.read_csv(url)
```

```python
#cleaning data
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y)
plt.show()    #get rid of random numbers above the graph
```

```python
df.job_posted_date.head()
```

```python
plt.plot(df.job_posted_date, df.job_posted_date)
plt.show()
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts = date_counts.sort_index()
date_counts
```

```python
plt.plot(date_counts.index, date_counts.values)
plt.show()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
df
```

```python
monthly_counts = df.job_posted_month.value_counts()    #monthly_counts is value counts of number of jobs posted per month
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts)     #index = job_posted_month & values = count
plt.show()
```

```python
job_counts = df.job_title_short.value_counts()
job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.show()
```

```python
job_counts = job_counts.sort_values(ascending = True)
plt.barh(job_counts.index, job_counts)
plt.show()
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title("Job Title Counts")
plt.ylabel("No. of Job Postings")
plt.xticks(rotation = 45,  ha = "right")    #rotate x-axis labels and horizontal alignment(ha) to right so that the labels doesn't stick to each other
plt.xlabel("Job Titles")
plt.show()
```
